```python
# a necessary library for writing out our transform rules mathematically
from sympy import symbols, lambdify, sin, cos, tanh, exp, log, Max

# our home-made warper 
import new_warper as warper
```

# A demonstration with a kernel

```python
# make instance of warper
demo1 = warper.warp_demo()
x1,x2 = symbols('x1 x2')

rule1 = x1*x1 - 4
rule2 = x2*x2 - 4

demo1.define_rule(rule1,rule2)
demo1.make_pts()
```

```python
slider1 = demo1.bokeh_transformation_slider()
```

```python
from ipywidgets import interact
interact(slider1, param=(0,400,1))
```

# A demo with a neural network

```python
# make instance of warper
demo2 = warper.warp_demo()
x1,x2 = symbols('x1 x2')

rule1 = tanh(x1 + tanh(x1 + x2))
rule2 = x1

demo2.define_rule(rule1,rule2)
demo2.make_pts()
```

```python
slider2 = demo2.bokeh_transformation_slider()
```

```python
from ipywidgets import interact
interact(slider2, param=(0,500,1))
```
